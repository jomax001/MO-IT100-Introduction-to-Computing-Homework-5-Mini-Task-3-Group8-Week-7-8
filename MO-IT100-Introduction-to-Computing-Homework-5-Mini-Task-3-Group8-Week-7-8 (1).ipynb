```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns

# General Disclaimer
print("Disclaimer: This analysis is based on the provided dataset and is intended for illustrative purposes only.")
print("The dataset is relatively small, and the findings may not be generalizable to a larger population.")
print("Interpret the results with caution and consider further investigation with more comprehensive data.")

# Load the dataset from the CSV file
file_path = 'Dataset.csv'
try:
    df = pd.read_csv(file_path)
except FileNotFoundError:
    current_dir = os.getcwd()
    print(f"Error: File not found at path: {file_path}. Make sure 'Dataset.csv' is in the same directory (currently: {current_dir}), or provide the correct path.")
    exit()

# Convert 'Adaptivity Level' to ordered category
df['Adaptivity Level'] = pd.Categorical(df['Adaptivity Level'], categories=['Low', 'Moderate', 'High'], ordered=True)

# Set a better style for the plots
sns.set(style="whitegrid")

# 1. Adaptivity vs. Network Type
plt.figure(figsize=(12, 7))  # Adjusted figure size
plt.suptitle('Adaptivity Levels by Network Type', fontsize=16, y=0.98)  # Overall Chart Title
plt.title('Distribution of Adaptivity for Each Network Type.  Small dataset - interpret with caution.\n\n\n', fontsize=14, color='gray', y=0.92)  # Subtitle
sns.countplot(x='Network Type', hue='Adaptivity Level', data=df, palette="viridis", order=['2G', '3G', '4G'])
plt.xlabel('Network Type', fontsize=14)
plt.ylabel('Number of Students', fontsize=14)
plt.legend(title='Adaptivity', loc='upper right')
plt.tight_layout(rect=[0, 0, 1, 0.95])  # Adjust layout for suptitle
plt.show()

# 2. Adaptivity vs. Device Type
plt.figure(figsize=(12, 7))  # Adjusted figure size
plt.suptitle('Adaptivity Levels by Device Type', fontsize=16, y=0.98)  # Overall Chart Title
plt.title('Distribution of Adaptivity for Each Device.  Small dataset - interpret with caution.\n\n\n', fontsize=14, color='gray', y=0.92)  # Subtitle
sns.countplot(x='Device', hue='Adaptivity Level', data=df, palette="muted")
plt.xlabel('Device Type', fontsize=14)
plt.ylabel('Number of Students', fontsize=14)
plt.legend(title='Adaptivity', loc='upper right')
plt.tight_layout(rect=[0, 0, 1, 0.95])  # Adjust layout for suptitle
plt.show()

# 3. Adaptivity vs. IT Student
plt.figure(figsize=(8, 7))  # Adjusted figure size
plt.suptitle('Adaptivity Levels by IT Student Status', fontsize=16, y=0.98)  # Overall Chart Title
plt.title('Distribution of Adaptivity for IT vs. Non-IT Students. Small dataset - interpret with caution.\n\n\n', fontsize=14, color='gray', y=0.92)  # Subtitle
sns.countplot(x='IT Student', hue='Adaptivity Level', data=df, palette="pastel")
plt.xlabel('IT Student', fontsize=14)
plt.ylabel('Number of Students', fontsize=14)
plt.legend(title='Adaptivity', loc='upper right')
plt.tight_layout(rect=[0, 0, 1, 0.95])  # Adjust layout for suptitle
plt.show()

# 4. Adaptivity vs. Financial Condition
plt.figure(figsize=(10, 7))  # Adjusted figure size
plt.suptitle('Adaptivity Levels by Financial Condition', fontsize=16, y=0.98)  # Overall Chart Title
plt.title('Distribution of Adaptivity for Each Financial Condition. Small dataset - interpret with caution.\n\n\n', fontsize=14, color='gray', y=0.92)  # Subtitle
sns.countplot(x='Financial Condition', hue='Adaptivity Level', data=df, palette="Set2", order=['Poor', 'Mid', 'Rich'])
plt.xlabel('Financial Condition', fontsize=14)
plt.ylabel('Number of Students', fontsize=14)
plt.legend(title='Adaptivity', loc='upper right')
plt.tight_layout(rect=[0, 0, 1, 0.95])  # Adjust layout for suptitle
plt.show()

# ========================= Recommendation =========================
print("\n===================================================== Recommendation ============================================================================\n")
print("Based on the provided dataset, stakeholders should prioritize improving internet access, particularly upgrading network infrastructure to 4G and \nbeyond, as this shows a positive association with student adaptivity. Targeted support and resources should be directed towards non-IT students and \nstudents with poor financial conditions, as IT proficiency and financial stability appear to correlate with higher adaptivity levels. \nAdditionally, consider the prevalence of mobile devices and optimize learning platforms accordingly, while also acknowledging the limitations of the \nsmall dataset and the need for further, more comprehensive research to validate these initial findings and inform more robust strategies.")
print("\n========================================================================================================================================================\n")

# We mainly used countplots in our analysis.
#
# A countplot is a simple way to count things and show them as bars.
# Think of counting how many people like different flavors of ice cream: chocolate,
# vanilla, and strawberry. A countplot would have a bar for each flavor, showing
# how many people chose it. (Waskom, 2021)
#
# In our charts, the x-axis shows the different groups we're studying: internet
# connection type (2G, 3G, 4G), the kind of device students use, whether they're
# studying IT, and their family's financial situation.  The y-axis counts the number
# of students in each of those groups.
#
# We also used a trick to split these groups further based on "Adaptivity Level"
# (Low, Moderate, High).  So, when we look at internet connection type, we can see
# how many students at *each* Adaptivity Level are using 2G, 3G, or 4G. This helps
# us see if certain Adaptivity Levels are more common with certain connection types.
#
# Reference:
# Waskom, M. L. (2021). seaborn: statistical data visualization. The Journal of Open Source Software. https://joss.theoj.org/papers/10.21105/joss.03021


```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from the CSV file
file_path = 'Dataset.csv'  # Replace with the actual path to your file
df = pd.read_csv(file_path)

# Filter for moderate to high adaptivity
df_adaptable = df[df['Adaptivity Level'].isin(['Moderate', 'High'])].copy()

# --- Data Preparation for Pie Charts ---

# 1. Internet Type:  (Crucial for online learning)
internet_type_counts = df_adaptable['Internet Type'].value_counts()

# 2. Device: (How they access the online classroom)
device_counts = df_adaptable['Device'].value_counts()

# 3. IT Student: (Do they have a background?)
it_student_counts = df_adaptable['IT Student'].value_counts()

# --- Plotting Pie Charts ---

fig, axes = plt.subplots(3, 1, figsize=(12, 24))  # Vertical subplots
fig.suptitle('\n\n\n\nKey Indicators of Moderate to High Adaptability in Online Learning', fontsize=20)

# Overall explanation text ABOVE the charts
fig.text(0.5, 0.95,
         "We analyzed key factors influencing adaptability in online learning among students with moderate to high levels. The pie charts below illustrate the distribution of these factors – specifically, internet type, device, and IT student status. The size of each slice shows its proportion within this adaptable group. For instance, a larger slice for 'Wifi' suggests a strong link between reliable internet and adaptability. Keep in mind these are indicators, not definitive causes.",
         ha="center", fontsize=14, wrap=True)

# 1. Internet Type Pie Chart
labels = internet_type_counts.index
sizes = internet_type_counts.values
colors = ['lightgreen', 'lightsalmon', 'lightblue']  # Distinct colors
explode = [0.1] + [0] * (len(labels) - 1)  # explode 1st slice, rest 0
axes[0].pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
axes[0].axis('equal')
axes[0].set_title('\nInternet Type\n(Proportion of Moderate/High Adaptability Students)', fontsize=16)

# Recommendations for Internet Type:
fig.text(0.1, 0.25,
         "Based on the internet type data, we recommend: \n"
         "- For Learning Institutions: Let's support students with less reliable internet. Think offline materials or data stipends.\n"
         "- For Students: Look for affordable, stable internet. Maybe find places with free Wi-Fi to study.\n"
         "- For Policymakers: Expanding affordable internet access programs can make a big difference.",
         ha="left", fontsize=12, wrap=True, bbox={'facecolor': 'white', 'alpha': 0.8, 'pad': 10})  # Added background box

# 2. Device Pie Chart
labels = device_counts.index
sizes = device_counts.values
colors = ['lightseagreen', 'plum', 'gold']  # Distinct colors
explode = [0.1] + [0] * (len(labels) - 1)  # explode 1st slice, rest 0
axes[1].pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
axes[1].axis('equal')
axes[1].set_title('\n\n\n\n\n\n\n\n\n\nDevice\n(Proportion of Moderate/High Adaptability Students)\n', fontsize=16)

# Recommendations for Device:
fig.text(0.1, 0.55,
         "Considering the device usage, we suggest: \n"
         "- For Learning Institutions: Make sure your platforms work well on all kinds of devices, even phones.\n"
         "- For Students: Use whatever device works best for you, and tweak the settings to help you focus.\n"
         "- For Families/Support Systems: If you can, help students get access to a computer or tablet – it really helps!",
         ha="left", fontsize=12, wrap=True, bbox={'facecolor': 'white', 'alpha': 0.8, 'pad': 10})  # Added background box

# 3. IT Student Pie Chart
labels = it_student_counts.index
sizes = it_student_counts.values
colors = ['skyblue', 'lightcoral']  # Distinct colors
explode = [0.1] + [0] * (len(labels) - 1)  # explode 1st slice, rest 0
axes[2].pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
axes[2].axis('equal')
axes[2].set_title('IT Student\n(Proportion of Moderate/High Adaptability Students)', fontsize=16)

# Recommendations for IT Student:
fig.text(0.1, 0.85,
         "From the IT student data, we propose: \n"
         "- For Learning Institutions: Offer some basic IT skills courses or workshops for students who haven't had much experience.\n"
         "- For Students: Use the IT resources and training that are out there. Even simple skills can make a difference.\n"
         "- For Curriculum Developers: Try to weave IT skills into the course content where it makes sense.",
         ha="left", fontsize=12, wrap=True, bbox={'facecolor': 'white', 'alpha': 0.8, 'pad': 10})  # Added background box

plt.tight_layout(rect=[0, 0.03, 1, 0.92])  # Adjust layout to make room for the title and recommendations
plt.show()  

# Explanation of Adaptability Factors in Online Learning

# Overview:
# These pie charts show the distribution of key factors among students with moderate to high adaptability in online learning, 
# based on our analysis of the dataset provided in [1]. We're examining how factors such as internet access, device choice, and IT skills relate to a student's
# ability to adjust to and succeed in an online learning environment. We used the provided dataset [1] to find the correlations between the factors.
# Remember that we are focusing on the features that most correlate to moderate to high adaptability level and not trying to say they are the cause.

# Internet Type Analysis:
# This pie chart breaks down the type of internet connection reported by adaptable students. Having reliable access is crucial
# for participating in online classes, submitting assignments, and engaging with resources.  Based on [1], students with Wifi show a strong association with adaptability,
# meaning students who reported having Wifi also reported moderate to high adaptability. This highlights the importance of Wifi in online learning environments.
# We recommend institutions support students lacking reliable internet through offline materials or stipends. Students should
# seek stable internet options, and policymakers should consider affordable access programs.

# Device Analysis:
# The device pie chart shows the different devices adaptable students use for online learning. This is important because the
# learning experience can vary significantly between devices. Based on the file [1], we observe the proportion of students who use mobile phones, computer, or Tab devices. It's important to ensure learning platforms function well across device types.
# We recommend institutions ensure platform accessibility on various devices, students optimize their device settings, and families/support
# systems provide access to computers/tablets when possible.

# IT Student Analysis:
# This pie chart shows the division between students who identify as IT students and those who do not. Based on [1], students who said they are IT students also
# have moderate to high adaptability in online learning, meaning having IT skills might
# provide an advantage in online learning, though it's not a requirement for success. We recommend
# institutions offer basic IT skills courses, students take advantage of IT resources, and curriculum developers integrate
# IT skills into course content.

# Conclusion:
# These charts, derived from our analysis of the dataset [1], highlight potential indicators of adaptability in online learning. By understanding
# these factors, we can work to create more equitable and effective online experiences. Remember that these factors only correlated with moderate to high adaptability level.
# Future research could explore these relationships in more depth.
```
